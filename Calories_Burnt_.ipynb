```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Data Collection & Processing


```python
# loading the data from csv file to a Pandas DataFrame
calories = pd.read_csv('/content/calories.csv')
```

```python
exercise_data = pd.read_csv('/content/exercise.csv')
```

Combining the two Dataframes

```python
calories_data = pd.concat([exercise_data, calories['Calories']], axis=1)
```

Converting the text data to numerical values

```python
calories_data.replace({"Gender":{'male':0,'female':1}}, inplace=True)
```

```python
calories_data.head()
```

Separating features and Target

```python
X = calories_data.drop(columns=['User_ID','Calories'], axis=1)
Y = calories_data['Calories']
print(X)
```

```python
print(Y)
```

Splitting the data into training data and Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

XGBoost Regressor

```python
# loading the model
model = XGBRegressor()
```

```python
# training the model with X_train
model.fit(X_train.select_dtypes(exclude=['object']), Y_train)
# Exclude columns with 'object' dtype to only use numeric/boolean data
```

Evaluation

Prediction on Test Data

```python
# Handle categorical features in X_train before training
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Identify categorical columns in X_train
categorical_cols_train = X_train.select_dtypes(include=['object']).columns

# Apply one-hot encoding to categorical columns in X_train
ohe_train = OneHotEncoder(handle_unknown='ignore', sparse=False)
X_train_encoded = pd.DataFrame(ohe_train.fit_transform(X_train[categorical_cols_train]))
X_train_encoded.columns = ohe_train.get_feature_names_out(categorical_cols_train)

# Drop original categorical columns and join encoded ones for X_train
X_train_processed = X_train.drop(categorical_cols_train, axis=1).join(X_train_encoded)

# Now train the model using the processed X_train
model.fit(X_train_processed, Y_train)

# ... (rest of your code for handling X_test remains the same)
```

```python
# Assuming 'model' is your trained XGBRegressor and 'X_test_processed' is your preprocessed test data
test_data_prediction = model.predict(X_test_processed)

print(test_data_prediction)
```

Mean Absolute Error

```python
mae = metrics.mean_absolute_error(Y_test, test_data_prediction)
```

```python
print("Mean Absolute Error = ", mae)
```
